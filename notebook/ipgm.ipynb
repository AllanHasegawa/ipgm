```python
from ia636 import *
from numpy import *
from scipy import *
from adpil import *
from pymorph.compat import *

def ipshow(f):
    #D = pylab.gcf().get_dpi()/2
    
    if len(f.shape) == 3:
        #figsize(f.shape[1]/72.,f.shape[2]/72)
        ff = zeros((f.shape[1], f.shape[2], 3), f.dtype)
        ff[:,:,0] = f[0,:,:]
        ff[:,:,1] = f[1,:,:]
        ff[:,:,2] = f[2,:,:]
    else:
        #figsize(f.shape[0]/72.,f.shape[1]/72.)
        ff = zeros((f.shape[0], f.shape[1], 3), f.dtype)
        ff[:,:,0] = f[:,:]
        ff[:,:,1] = f[:,:]
        ff[:,:,2] = f[:,:]
    imshow(ff)

f = adread('img6.jpg')

ipshow(f)
```

```python

def rgb2hsl(f):
    m = f / 255.
    
    dim,h,w = m.shape
    
    for x in range(h):
        for y in range(w):
            r,g,b = m[:,x,y]
            cmax = max(r,g,b)
            cmax_i = argmax((r,g,b))
            cmin = min(r,g,b)
            
            h = (cmax+cmin)/2
            s = h
            l = h
            if cmax == cmin:
                s = 0
                h = 0
            else:
                d = cmax-cmin
                
                s = d / (cmax+cmin)
                if l > 0.5:
                    s = d / (2-cmax-cmin)
                    
                if cmax_i == 0:
                    if g < b:
                        h = (g - b) / d + 6
                    else:
                        h = (g - b) / d + 0
                elif cmax_i == 1:
                    h = (b - r) / d + 2
                elif cmax_i == 2:
                    h = (r - g) / d + 4
                h /= 6;

            m[0,x,y] = h
            m[1,x,y] = s
            m[2,x,y] = l
    return m
    

def segmentSensorLED(f, ID):
    hs = 13
    ws = 13
    xi = 520
    yi = 94
    wss = 9

    xii = xi
    yii = yi+ID*(wss+ws)
    g = f[:,xii:xii+hs,yii:yii+ws]
    
    return g
```

```python
def detectSingleSensorLED(f,ID,t = 100):
    led = segmentSensorLED(f,ID)
    led_l = mmsubm(led[0],led[2]) > 10
    return sum(led_l) > t

for i in range(11):
    print "Sensor ID[", i, "]: ", detectSingleSensorLED(f,i)
    
```

```python
def segmentActuatorLED(f, ID):
    hs = 22
    ws = 21
    xi = 555
    yi = 90
    wss = 1

    xii = xi
    yii = yi+ID*(wss+ws)
    g = f[:,xii:xii+hs,yii:yii+ws]
    
    return g

ipshow(segmentActuatorLED(f,0))
```

```python
def detectSingleActuatorLED(f,ID,t = 100):
    led = segmentActuatorLED(f,ID)
    led_l = mmsubm(led[0],led[2]) > 10
    return sum(led_l) > t

for i in range(8):
    print "Actuator ID[", i, "]: ", detectSingleActuatorLED(f,i)
```
