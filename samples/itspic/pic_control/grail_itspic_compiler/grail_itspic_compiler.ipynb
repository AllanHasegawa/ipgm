```python
import exceptions

def parse_sup(file_name):
    # this set will contain all nodes
    ### set as unique values will be inserted;
    ### list is used because I want an index
    # each node is a unique INTEGER
    nodes = []
    
    # this set will contain all actions
    # each action is a unique String
    actions = []
    
    # this set will contain all transitions
    # each transition is a tuple of (origin node, action, destiny node)
    transitions = []
    
    #TODO hardcoded File name
    f = open(file_name, 'r')
    
    for line in f:
        s = line.replace('\n', '').split(' ')
        if len(s) != 3:
            raise Exception('Input line dont have 3 words')
        try:
            origin_node = int(s[0])
            try:
                nodes.index(origin_node)
            except ValueError:
                nodes.append(origin_node)
                pass
        except exceptions.ValueError:
            if s[0] == '(START)':
                origin_node = -1
            else:
                raise Exception('Cant understand origin node, found: %s' % s[0])
            
        try:
            dst_node = int(s[2])
            try:
                nodes.index(dst_node)
            except ValueError:
                nodes.append(dst_node)
                pass
        except exceptions.ValueError:
            if s[2] == '(FINAL)':
                dst_node = -2
            else:
                raise Exception('Cant understand dst node, found: %s' % s[2])
        
        action = s[1]
        if action == '-|' or action == '|-':
            continue
        try:
            actions.index(action)
        except ValueError:
            actions.append(action)
            pass
        
        transitions.append( (origin_node,action,dst_node) )

    f.close()
    
    return (nodes,actions,transitions)
```

```python
s0_filename = 'SUP_ELEVATOR_FITTER.fm'

print "Loading:", s0_filename
s0_nodes,s0_actions,s0_edges = parse_sup(s0_filename)

print
print 'Number of nodes: ', len(s0_nodes)
print 'Number of actions:', len(s0_actions)
print 'Number of transitions:', len(s0_edges)
print

for ai in range(len(s0_actions)):
        print '#define EVENT_%s %d' % (s0_actions[ai].upper(),ai)

print
print

for a in s0_actions:
    if not a.startswith('end_'):
        print 'void %s(void) {\n}\n' % a
        print 'void update_%s(void) {\n}\n' % a
```
