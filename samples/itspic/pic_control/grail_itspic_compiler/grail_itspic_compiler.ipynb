```python
import exceptions

def parse_sup(file_name):
    # this set will contain all nodes
    ### set as unique values will be inserted;
    ### list is used because I want an index
    # each node is a unique INTEGER
    nodes = []
    
    # this set will contain all actions
    # each action is a unique String
    actions = []
    
    # this set will contain all transitions
    # each transition is a tuple of (origin node, action, destiny node)
    transitions = []
    
    #TODO hardcoded File name
    f = open(file_name, 'r')
    
    for line in f:
        s = line.replace('\n', '').split(' ')
        if len(s) != 3:
            raise Exception('Input line dont have 3 words')
        try:
            origin_node = int(s[0])
            try:
                nodes.index(origin_node)
            except ValueError:
                nodes.append(origin_node)
                pass
        except exceptions.ValueError:
            if s[0] == '(START)':
                origin_node = -1
            else:
                raise Exception('Cant understand origin node, found: %s' % s[0])
            
        try:
            dst_node = int(s[2])
            try:
                nodes.index(dst_node)
            except ValueError:
                nodes.append(dst_node)
                pass
        except exceptions.ValueError:
            if s[2] == '(FINAL)':
                dst_node = -2
            else:
                raise Exception('Cant understand dst node, found: %s' % s[2])
        
        action = s[1]
        if action == '-|' or action == '|-':
            continue
        try:
            actions.index(action)
        except ValueError:
            actions.append(action)
            pass
        
        transitions.append( (origin_node,action,dst_node) )
        
    # I'll makes nodes linear increasing, starting with 0    
    new_transitions = []
    
    for i in transitions:
        new_transitions.append((nodes.index(i[0]), i[1], nodes.index(i[2])))

    f.close()
    
    return (len(nodes),actions,new_transitions)
```

```python
s_filenames = ['SUP_ELEVATOR_FITTER.fm', 'SUP_FITTER_FLOOR.fm', 'SUP_IBOX_FLOOR.fm', 'SUP_IPALLET_ELEVATOR.fm' ]

sups = []

for fn in s_filenames:
    print "Loading:", fn
    t = parse_sup(fn)
    sups.append((fn,t[0],t[1],t[2]))

for s in sups:
    

    print
    print "FileName: ", s[0]
    print 'Number of nodes: ', s[1]
    print 'Number of actions:', len(s[2])
    print 'Number of transitions:', len(s[3])
    print s[3]
    print "\n"*2
```

```python
# actions controlable
actions_c = set()
# actions non controlable
actions_n = set()
# all actions
actions = set()
for i in sups:
    for j in i[2]:
        actions.add(j)
        if (j.startswith('end_')):
            actions_n.add(j)
        else:
            actions_c.add(j)

actions = tuple(actions)
actions_c = tuple(actions_c)
actions_n = tuple(actions_n)

#print actions[10]
#print actions.index('close_floor')

def print_define_events():
    r = ""
    for i in range(len(actions_c)):
        r += '#define EVENT_C_%s %d\n' % (actions_c[i].upper(),i)
    for i in range(len(actions_n)):
        r += '#define EVENT_N_%s %d\n' % (actions_n[i].upper(),len(actions_c)+i)
    
    r += '\n'*2
    
    for i in range(len(actions_c)):
        r += '#define DES_EVENT_%s %s\n' % (actions_c[i].upper(),print_set_des(actions_c[i]))
        
    r += '\n'*2
    
    for i in range(len(actions_c)):
        r += '#define IS_DES_EVENT_%s %s\n' % (actions_c[i].upper(),print_is_des(actions_c[i]))
    return r
    

def print_des_vars():
    s = ""
    for i in range(0,len(actions_c),8):
        s = s + ('bybit des%d;' % (i/8)) + "\n"
    return s
        
def print_set_des(a,b = True):
    a_id = actions_c.index(a)
    a_bank = (a_id / 8)
    if b:
        a_p = (a_bank,(a_id%8),1)
    else:
        a_p = (a_bank,(a_id%8),0)
    return 'des%d.bits.b%d = %d;' % a_p

def print_clear_des():
    s = ""
    for i in range(0,len(actions_c),8):
        s = s + ('des%d.byte = 0;' % (i/8)) + "\n"
    return s

def print_is_des(a):
    a_id = actions_c.index(a)
    a_bank = (a_id / 8)
    return '(des%d.bits.b%d == 1)' % (a_bank,a_id%8)
    
    
#print actions
#print
#print actions_c, len(actions_c)
#print
#print actions_n
#print print_define_events()
#print
#print print_des_vars()
#print "//testando set close_floor"
#print print_set_des('close_floor')
#print "//testando set move_1"
#print print_set_des('move_1')
#print "//testando set move_2(F)"
#print print_set_des('move_2', False)
```

```python


def print_sup_response_step():
    r = ""
    
    #r = 'unsigned char events_n_todo[20];\n'
    #r += 'unsigned char events_n_todo_count;\n\n'
    r += 'void sup_response_step() {\n'

    r += 'unsigned char s0_b = s0_s;\n'
    r += 'unsigned char s1_b = s1_s;\n'
    r += 'unsigned char s2_b = s2_s;\n'
    r += 'unsigned char s3_b = s3_s;\n'
    r += 'if (events_n_todo_count > 40) {\n'
    r += 'SERIAL_TX(243);\n'
    r += '}\n'
    r += 'unsigned char i = 0;\n'
    r += 'for (; i < events_n_todo_count;i++) {\n'
    r += 'const unsigned char a_event = events_n_todo[i];\n'
    for s_i in range(len(sups)):
        s = sups[s_i]
        
        f1 = True
        for i in range(s[1]):
            l_n = [x for x in s[3] if x[0] == i and x[1].startswith('end_')]
            if len(l_n):
                if f1:
                    f1 = False
                    r += 'if (s%d_s == %d) {\n' % (s_i,i)
                else:
                    r += 'else if (s%d_s == %d) {\n' % (s_i,i)
                
                f2 = True
                for l in l_n:
                    if f2:
                        f2 = False
                        r += '\tif (a_event == %s) {\n' % ('EVENT_N_%s' % l[1].upper())
                    else:
                        r += '\telse if (a_event == %s) {\n' % ('EVENT_N_%s' % l[1].upper())
                    r += '\t\ts%d_s = %d;\n' % (s_i,l[2])
                    r += '\t}\n'
                r += '}\n'
    r += '}\n'
    r += 'events_n_todo_count = 0;\n'
    r += 'if (s0_b != s0_s) {\n'
    r += 'SERIAL_TX(s0_s+200);\n'
    r += '}\n'
    r += 'if (s1_b != s1_s) {\n'
    r += 'SERIAL_TX(s1_s);\n'
    r += '}\n'
    r += 'if (s2_b != s2_s) {\n'
    r += 'SERIAL_TX(s2_s+10);\n'
    r += '}\n'
    r += 'if (s3_b != s3_s) {\n'
    r += 'SERIAL_TX(s3_s+50);\n'
    r += '}\n'
    r += '}\n\n'
    
    return r

#print print_sup_response_step()
```

```python
def print_s_des(sup_i):
    sup = sups[sup_i]
    r = ""    
    t = sup[3]
    # all states
    first = True
    for i in range(sup[1]):
        l_n = [x[1] for x in sup[3] if x[0] == i and not x[1].startswith('end_')]
        #print i, l_n
        t_a = [x for x in sup[2] if not x.startswith('end_')]
        for j in l_n:
            t_a.remove(j)            
        #print i, 't_a', t_a
        if first:
            r += "if (s%d_s == %d) {\n" % (sup_i,i)
            first = False
        else:
            r += "else if (s%d_s == %d) {\n" % (sup_i,i)
        
        for j in t_a:
            r += 'DES_EVENT_' + j.upper() + ';\n'
        r += '}\n'
    return r
    

def print_des_step():
    r = ""
    r += 'void des_step(void) {\n';
    r += print_clear_des()
    for i in range(len(sups)):
        r += print_s_des(i)
    r += '}\n\n'
    return r

#print print_des_step()
```

```python
def print_dec_step():
    r = ""
    r += "unsigned char decisions[%d];\n" % len(actions_c)
    r += "unsigned char decisions_count;\n"
    r += 'unsigned char decided_action;\n'
    
    r += "void decisions_step() {\n"
    r += 'decisions_count = 0;\n'
    
    for i in actions_c:
        r += 'if (!IS_DES_EVENT_%s) {\n' % (i.upper())
        r += '\tdecisions[decisions_count++] = %s;\n' % ('EVENT_C_%s' % i.upper())
        r += '}\n'
                
    r += 'if (decisions_count > 0) {\n'
    r += 'decided_action = decisions[rand() % decisions_count];\n'
    r += '}\n'
    r += 'else {\n'
    r += 'decided_action = 255;\n'
    r += '}\n'
    
    r += '}\n\n'
    return r

#print print_dec_step()
```

```python
def print_sup_advance_step():
    r = ""
    
    r += 'void sup_advance_step() {\n'
    r += 'unsigned char s0_b = s0_s;\n'
    r += 'unsigned char s1_b = s1_s;\n'
    r += 'unsigned char s2_b = s2_s;\n'
    r += 'unsigned char s3_b = s3_s;\n'
    
    f1 = True
    for i in actions_c:
        if f1:
            r += 'if (decided_action == EVENT_C_%s) {\n' % i.upper()
            f1 = False
        else:
            r += 'else if (decided_action == EVENT_C_%s) {\n' % i.upper()
            r += 'tmembank.bits.b1 = 0;\n'
        
        for s_i in range(len(sups)):
            sup = sups[s_i]
            
            if len([x for x in sup[2] if x == i]):
                f2 = True
                for n in range(sup[1]):
                    l_a = [x for x in sup[3] if x[0] == n and x[1] == i]
                    if len(l_a):
                        if f2:
                            r += 'if (s%d_s == %d) {\n' % (s_i,n)
                            f2 = False
                        else:
                            r += 'else if (s%d_s == %d) {\n' % (s_i,n)
                        r += 's%d_s = %d;\n' % (s_i,l_a[0][2])
                        r += 'tmembank.bits.b1 = 1;\n'
                        r += '}\n'
        
        r += 'if (tmembank.bits.b1 == 1){\n'
        r += '%s();\n' % i
        r += '}\n'
        r += '}\n\n'
    
    r += 'if (s0_b != s0_s) {\n'
    r += 'SERIAL_TX(s0_s+200);\n'
    r += '}\n'
    r += 'if (s1_b != s1_s) {\n'
    r += 'SERIAL_TX(s1_s);\n'
    r += '}\n'
    r += 'if (s2_b != s2_s) {\n'
    r += 'SERIAL_TX(s2_s+10);\n'
    r += '}\n'
    r += 'if (s3_b != s3_s) {\n'
    r += 'SERIAL_TX(s3_s+50);\n'
    r += '}\n'
    r += '}\n'
    return r

#print print_sup_advance_step()
```

```python
def mount_file():
    with open('_header.c', 'r') as x:
        f_header = x.read()
    with open('_defines.c', 'r') as x:
        f_defines = x.read()
    with open('_global_variables.c', 'r') as x:
        f_global_variables = x.read()
    with open('_sequence_operations.c', 'r') as x:
        f_sequence_operations = x.read()
    with open('_interrupt.c', 'r') as x:
        f_interrupt = x.read()
    with open('_main.c', 'r') as x:
        f_main = x.read()
        
    f = ''
    
    f += f_header
    
    f += f_defines
    
    f += print_define_events()
    
    #f += print_des_vars
    f += f_global_variables
    
    f += f_sequence_operations
    
    f += print_sup_response_step()
 
    f += print_des_step()
    
    f += print_dec_step()
    
    f += print_sup_advance_step()
    
    f += f_interrupt
    
    f += f_main
    
    with open('main.c', 'w') as myFile:
        myFile.write(f)

mount_file()
```
